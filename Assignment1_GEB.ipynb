## Handling missing data, encoding categorical columns, and normalizing the data 

```python
dataset.info()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
```

```python
dataset['Age']=imputer.fit_transform
```

```python
dataset['Cabin']=imputer.fit_transform
```

```python
dataset['Embarked']=imputer.fit_transform
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset[['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']], drop_first=True)
```

## Splitting the dataset using Stratified Sampling technique

```python
from sklearn.model_selection import train_test_split
strat_training_set, strat_test_set=train_test_split(dataset, test_size=0.2, stratify=dataset['Age'],
                                                   random_state=6)
```
